```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from datetime import datetime
```

```python
def log_message(message):
    timestamp = '%Y-%h-%d-%H:%M:%S'
    now = datetime.now()
    str_format = now.strftime(timestamp)
    with open('game_log.txt', 'a') as f:
        f.write(str_format + "_" + message + '\n')
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_video_games'
page = requests.get(url).content
soup = BeautifulSoup(page, 'html.parser')
```

```python
t_data = soup.find_all('table', class_='wikitable sortable plainrowheaders')
# t_data = soup.find_all('table')

# print(type(t_data))
t_data
rows = t_data[0].find_all('tr')
```

```python
table = []
for row in rows:
    data = row.find_all(['td', 'th'])
    print(data)
    print('---------------\n',data[0])
    print('---------------\n', data[0].text)
    break
```

```python
table = []
for row in rows:
    data = row.find_all(['td', 'th'])
    t_row = [each.text.strip() for each in data]
    table.append(t_row) #1st row: headers
table[10]
# t = [each.text.strip() for each in table[9]]
# t
rows
```

```python
cols = ['Title', 'Sales', 'Series', 'Platform', 'ReleaseDate', 'Developer(s)', 'Publisher(s)', 'Ref']
df = pd.DataFrame(table[1:], columns=cols)
# df = df.iloc[:, :-1]
df = pd.DataFrame(table)
df
```

```python
# save to db;
import sqlite3 as sq

conn = sq.connect('games.db')
table_name = 'top_games'

df.to_sql(table_name, conn, if_exists='replace', index=False)
```
