```python
import pandas as pd
```

```python
from datetime import datetime

def log(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S'
    now = datetime.now()
    current_time = now.strftime(timestamp_format)
    with open('log.txt', 'a') as f:
        f.write(current_time+', '+message + '\n')
```
